## Full-time Employment VS Education

The graph below shows the correlation between the levels of education and being employed full-time. The graph uses linear regression to show the positive relationship between the level of education and the number of people who are in full-time employment. This graph shows that the higher your level of education, the greater chance you have of getting into full-time employment. (To refrence the points on the grpah, use the numbers shown above) 

```python
from scipy.interpolate import *
from matplotlib.pyplot import *
from numpy import *
x = data['FormalEducation'].value_counts()
y = data[data['EmploymentStatus'] == 'Employed full-time']['FormalEducation'].value_counts()
plt.figure(figsize=(15,8))
p1 = polyfit(x,y,1)
plt.scatter(x,y,color='k')
plot(x,polyval(p1,x),'r-')
plt.xlabel('Formal Education')
plt.ylabel('Employed Full-Time')
plt.show()
```

## Part-time Employmeny VS Education

This graph enforces the gaph above by supporting the fact that the higher level of education you have increases your chances of employment. (To refrence the points on the graph use the numbers displayed above)

```python
from scipy.interpolate import *
from matplotlib.pyplot import *
from numpy import *
x = data['FormalEducation'].value_counts()
y = data[data['EmploymentStatus'] == 'Employed part-time']['FormalEducation'].value_counts()
plt.figure(figsize=(15,8))
p1 = polyfit(x,y,1)
plt.scatter(x,y,color='k')
plot(x,polyval(p1,x),'r-')
plt.xlabel('Formal Education')
plt.ylabel('Employed Part-Time')
plt.show()
```
