## Parents Impact on Education Success 

After looking at parents education having an impact on their childrens views on the importance of education we wanted to see if it had an impact on their education success. The bar chart and graph below displays the results of people who have achieved a Bachelo's Degree within each category of highest educated parents. From the results you can see that their is an almost steady incline with the rising level of educated parents and then their is a huge spike when a parent has completed a Bachelor's Degree. According to the results being displayed below, the people with parents which have a high level of education have more success in achieving higher level education.

```python
data[data['FormalEducation'] == "Bachelor's degree" ]["HighestEducationParents"].value_counts().plot(kind='bar', figsize=(9,7))
```

```python
x = data['HighestEducationParents'].value_counts()
y = data[data['FormalEducation'] == "Bachelor's degree" ]["HighestEducationParents"].value_counts()
plt.figure(figsize=(15,8))
p1 = polyfit(x,y,1) # linear fit
p2 = polyfit(x,y,2) # quadratic fit
p3 = polyfit(x,y,3) # cubic fit 
plt.scatter(x,y,color='k')
plot(x,polyval(p1,x),'r-')
plot(x,polyval(p2,x),'b-')
plot(x,polyval(p3,x),'g--')
plt.xlabel('Parents Education')
plt.ylabel('Respondents with Bachelors Degree')
plt.show()
```
